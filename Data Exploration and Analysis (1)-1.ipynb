### **Exploratory Analysis**
"Exploratory Analysis focuses on exploring and understanding the data without preconceived hypotheses. It involves visualizations, summary statistics, and data profiling techniques to uncover patterns, relationships, and interesting features. It helps generate hypotheses for further analysis"

```python
import pandas as pd

file_path = r'C:\Users\Sanketh\Pictures\Project 3\Music GEnre dataset\hit_songs\Hit Songs\spotify_hits_dataset_complete.csv'
spotify_data = pd.read_csv(file_path, delimiter='\t')

print(spotify_data.info())
print(spotify_data.head())
```

```python
# Summary statistics
summary_statistics = spotify_data.describe()
print(summary_statistics)

# Checking for missing values
missing_values = spotify_data.isnull().sum()
print(missing_values)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

def plot_numerical_features(data, columns):
    plt.figure(figsize=(15, 10))
    for i, column in enumerate(columns, 1):
        plt.subplot(2, 2, i)
        sns.histplot(data[column], bins=30, kde=True)
        plt.title(f'Distribution of {column}')
        plt.xlabel(column)
        plt.ylabel('Frequency')
    plt.tight_layout()
    plt.show()

numerical_columns = ['popularity', 'danceability', 'energy', 'loudness']
plot_numerical_features(spotify_data, numerical_columns)
```

The histograms provide insights into the distribution of key numerical features in your dataset:

Popularity: The distribution is somewhat left-skewed, indicating a larger number of songs have a popularity score below the average. There's a notable peak around the 50-60 range.

Danceability: This feature shows a right-skewed distribution, suggesting most songs in the dataset are relatively high in danceability, with a peak around 0.7-0.8.

Energy: The energy of songs is more evenly distributed but slightly skewed towards higher values, with a peak around 0.7-0.8.

Loudness: The distribution of loudness is left-skewed, with most songs having loudness levels clustered around -6 to -4 dB.

```python
plt.figure(figsize=(12, 6))

categorical_columns = ['song_type', 'explicit']

for i, column in enumerate(categorical_columns, 1):
    plt.subplot(1, 2, i)
    sns.countplot(y=spotify_data[column])
    plt.title(f'Count of {column}')
    plt.xlabel('Count')
    plt.ylabel(column)

plt.tight_layout()
plt.show()

```

The count plots provide insights into the distribution of two categorical features in your dataset:

Song Type: The majority of songs are classified as "Solo" performances, with a smaller number categorized as "Collaboration". This indicates that solo performances are more prevalent in this dataset.

Explicit Content: A significant portion of songs are marked as non-explicit, suggesting that explicit content is less common in the dataset.

### **Diagnostic Analysis**
"Diagnostic analysis aims to understand the cause-and-effect relationships within the data. It investigates the factors or variables that contribute to specific outcomes or behaviors. Techniques such as regression analysis or correlation analysis are commonly used."

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

file_path = r'C:\Users\Sanketh\Pictures\Project 3\Music GEnre dataset\hit_songs\Hit Songs\spotify_hits_dataset_complete.csv'
spotify_data = pd.read_csv(file_path, delimiter='\t')

numeric_data = spotify_data.select_dtypes(include=[float, int])

correlation_matrix = numeric_data.corr()

plt.figure(figsize=(12, 10))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f")
plt.title("Correlation Matrix of Spotify Dataset Numeric Features")
plt.show()
```

Popularity and Available Markets: There is a significant positive correlation (0.76) between popularity and num_available_markets, suggesting that songs available in more markets tend to be more popular.

Danceability and Valence: There's a positive correlation (0.59) between danceability and valence, indicating that songs which are more danceable tend to also be more positive (have higher valence).

Energy and Loudness: There is a very high positive correlation (0.73) between energy and loudness, which is intuitive as louder songs are often perceived as more energetic.

Energy and Acousticness: These two have a negative correlation (-0.42), implying that songs with higher energy are less likely to be acoustic.

Speechiness and Explicit Content: There is a moderate positive correlation (0.44) between speechiness and explicit, suggesting that songs with more spoken words are more likely to contain explicit content.

Instrumentalness and Loudness: There's a negative correlation (-0.38) between instrumentalness and loudness, indicating that more instrumental tracks might tend to be quieter.

No Strong Correlation with Key and Mode: Musical key and mode do not show strong correlations with other variables, indicating that they do not have a linear relationship with the popularity or other audio features of songs.

Weak Relationship with Track Number: The track number within an album does not show a strong correlation with the popularity or other features, which suggests that it might not be a significant factor in the characteristics of a song.

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score

X = spotify_data[['danceability', 'energy', 'loudness']]
y = spotify_data['popularity']

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

model = LinearRegression()
model.fit(X_train, y_train)

y_pred = model.predict(X_test)

mse = mean_squared_error(y_test, y_pred)
r2 = r2_score(y_test, y_pred)

print("Mean Squared Error:", mse)
print("R Score:", r2)
```

The R value is very low, indicating that the model does not explain much of the variance in the popularity variable. This suggests that while danceability, energy, and loudness might have some relationship with popularity, they are not strong predictors on their own.

### Key Takeaways



- **Market Availability's Influence on Popularity**: A strong positive correlation between the number of available markets and song popularity suggests that broader distribution is crucial for a song's success.

- **Musical Features and Song Popularity**: The relationship between popularity and audio features like danceability, energy, and loudness is not as strong as one might expect, with a very low R value in the regression analysis indicating that these features alone do not significantly predict popularity.

- **Energetic and Loud**: The high positive correlation between energy and loudness implies that these two attributes tend to go hand in hand, with more energetic songs typically being louder.

- **Danceability and Positivity**: The positive correlation between danceability and valence (positiveness of a track) indicates a trend where more danceable songs tend to be more upbeat and positive in nature.

- **Explicit Content and Speech**: There is a moderate correlation between explicit content and speechiness, indicating that tracks with more spoken words are more likely to be explicit.

- **Distribution Skewness**: The histograms show that several features like popularity, danceability, and loudness have skewed distributions, which could imply the presence of outliers or specific trends within the dataset that merit further investigation.
